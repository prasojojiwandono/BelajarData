## merging

```python
dfa = pd.DataFrame([['a',1],['b',2]],columns=['col1','number'])
dfb = pd.DataFrame([['c',3,'tiger'],['d',4,'lion']],columns=['letter','number','animal'])
```

```python
dfa
```

```python
dfb
```

```python
pd.concat([dfa,dfb])
```

```python
pd.concat([dfa,dfb],axis=1)
```

```python
pd.concat(
    [dfa,dfb],
    axis = 0,
    ignore_index=True
)
```

```python
df_ee = pd.DataFrame({
    'city':['jakarta','bogor','tangerang','depok','bekasi'],
    'temp':[34,54,34,64,23]
})
df_rr = pd.DataFrame({
    'city':['bogor','tangerang','depok','jakarta'],
    'humid':[34,34,64,23]
})
```

```python
df_ee
```

```python
df_rr
```

```python
pd.merge(df_ee,df_rr, on='city')
```

```python
pd.merge(df_ee,df_rr, on='city', how='outer')
```

```python
pd.merge(df_ee,df_rr, on='city', how='left')
```
