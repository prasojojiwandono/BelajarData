```python
import plotly.graph_objects as go
import plotly
import pandas as pd

```

```python
df_saham = pd.read_excel('/home/prasojo/belajar/BelajarData/dataset/buat_test_saham_aja.xlsx')
```

```python
df_saham.head()
```

```python
df_saham = df_saham.reindex(index=df_saham.index[::-1])
```

```python
df_saham = df_saham.set_index(pd.DatetimeIndex(df_saham['date'].values))
```

```python
df_saham
```

```python
figure = go.Figure(
    data =[
        go.Candlestick(
        x = df_saham.index,
        low = df_saham['Low'],
        high = df_saham['High'],
        close = df_saham['Close'],
        open = df_saham['Open'],
        increasing_line_color = 'green',
        decreasing_line_color = 'red',
        )
    ],
    layout=go.Layout(
        title=go.layout.Title(text="AALI")
    )
    
)

# figure.update_layout(
#     title = 'AALI'
# )

figure.show()
go.Candlestick()



```

```python
figure.write_image(file = '/home/prasojo/belajar/BelajarData/dataset/AALI.png',
                  width = 1600,
                  height = 1000)
# plotly.offline.plot(figure, filename='/home/prasojo/belajar/BelajarData/dataset/AALI.png')
# figure.to_image(format="png")
```
